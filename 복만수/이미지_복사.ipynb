# 폴더 안 이미지 복사

```python
import shutil
from pathlib import Path

folder = Path(r"C:\deep\data1\train")
src = folder / "정답-전입신고서1.jpg"

if not src.exists():
    raise FileNotFoundError(src)

for i in range(1, 1001):
    dst = folder / f"정답-전입신고서{i}.jpg"
    # i=1이면 원본과 동일 이름이므로 복사 안 함(덮어쓰기 방지)
    if i == 1:
        continue
    shutil.copy2(src, dst)

print("done")
```

# 폴더 안 이름 바꾸기

```python
from pathlib import Path

folder = Path(r"C:\Users\user\Desktop\2차 프로젝트\민원 서류 예시\새 폴더")

for file in folder.glob("*.jpg"):
    if file.name.startswith("학습-전입신고서"):
        new_name = file.name.replace("학습-전입신고서", "학습-손글씨체-전입신고서", 1)
        new_path = file.with_name(new_name)
        file.rename(new_path)
        print(f"{file.name} -> {new_name}")
```

# 폴더에 있지만 csv에는 없는거 삭제하기

```python
import os
import pandas as pd

folder_path = r"C:\Users\user\Desktop\2차 프로젝트\민원 서류 예시\데이터_복사본\텍스트_손글씨체_4000장\val_4000"
csv_path = r"C:\Users\user\Desktop\2차 프로젝트\민원 서류 예시\데이터_복사본\labels_4000.csv"
output_csv_path = r"C:\Users\user\Desktop\2차 프로젝트\민원 서류 예시\데이터_복사본\val_labels_4000.csv"

# 폴더 안 파일명 목록 가져오기 (.jpg 포함 그대로)
file_names = set(
    f for f in os.listdir(folder_path)
    if os.path.isfile(os.path.join(folder_path, f))
)

# CSV 읽기
df = pd.read_csv(csv_path)

# 첫 번째 컬럼이 파일명이라고 가정
first_col = df.columns[0]

# 공백 제거
df[first_col] = df[first_col].astype(str).str.strip()

# 폴더에 있는 파일명과 같은 행만 남기기
df_matched = df[df[first_col].isin(file_names)]

# 새 CSV로 저장
df_matched.to_csv(output_csv_path, index=False, encoding="utf-8-sig")

print("원본 행 수:", len(df))
print("남은 행 수:", len(df_matched))
print("삭제된 행 수:", len(df) - len(df_matched))
print("저장 완료:", output_csv_path)
```

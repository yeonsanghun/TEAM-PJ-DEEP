```python
import os, shutil, random

def dataset_split(train_ratio, val_ratio, seed=42):
    source_root = r'C:\Users\user\Desktop\2차 프로젝트\민원 서류 예시\데이터\텍스트_1000장'
    source_dir  = os.path.join(source_root)

    train_dir = r'C:\Users\user\Desktop\2차 프로젝트\민원 서류 예시\데이터\train_1000'
    val_dir   = r'C:\Users\user\Desktop\2차 프로젝트\민원 서류 예시\데이터\val_1000'
    test_dir  = r'C:\Users\user\Desktop\2차 프로젝트\민원 서류 예시\데이터\test_1000'

    for d in (train_dir, val_dir, test_dir):
        os.makedirs(d, exist_ok=True)

    file_list = [f for f in os.listdir(source_dir) if os.path.isfile(os.path.join(source_dir, f))]

    random.seed(seed)          # ✅ 추가 1: 시드 고정
    random.shuffle(file_list)  # ✅ 추가 2: 이제 매번 동일하게 섞임

    n_total = len(file_list)
    train_cnt = int(n_total * train_ratio)
    val_cnt   = int(n_total * val_ratio)

    print("total:", n_total, "train_cnt:", train_cnt, "val_cnt:", val_cnt, "test_cnt:", n_total-train_cnt-val_cnt)

    for cnt, file_name in enumerate(file_list):
        src = os.path.join(source_dir, file_name)

        if cnt < train_cnt:
            dst = os.path.join(train_dir, file_name)
        elif cnt < train_cnt + val_cnt:
            dst = os.path.join(val_dir, file_name)
        else:
            dst = os.path.join(test_dir, file_name)

        shutil.move(src, dst)

dataset_split(0.7, 0.15, seed=42)
```

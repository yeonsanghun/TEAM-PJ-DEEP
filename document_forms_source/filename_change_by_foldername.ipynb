```python
import os
import re

root_dir = r"C:\potenup3\potenup3-pj02-Project\서식\[별지+제1호서식]+확정일자+신청서(상가건물+임대차계약서상의+확정일자+부여+및+임대차+정보제공에+관한+규칙)"

for root, dirs, files in os.walk(root_dir):
    folder_name = os.path.basename(root)

    # 🔥 공백을 _ 로 변환
    safe_folder_name = folder_name.replace(" ", "_")

    for file in files:
        if not file.lower().endswith(".jpg"):
            continue

        try:
            # 앞의 3자리 숫자 추출
            match = re.match(r"(\d{3})", file)
            if not match:
                continue

            number_part = match.group(1)

            new_filename = f"{number_part}-{safe_folder_name}.jpg"

            old_path = os.path.join(root, file)
            new_path = os.path.join(root, new_filename)

            # 이미 존재하면 스킵
            if os.path.exists(new_path):
                print(f"이미 존재: {new_filename}")
                continue

            os.rename(old_path, new_path)
            print(f"변경 완료: {file} → {new_filename}")

        except Exception as e:
            print(f"에러: {file} → {e}")
```

```python
# 폴더 이름 변경
import os

root_dir = r"C:\potenup3\potenup3-pj02-Project\서식\[별지+제1호서식]+확정일자+신청서(상가건물+임대차계약서상의+확정일자+부여+및+임대차+정보제공에+관한+규칙)"

for root, dirs, files in os.walk(root_dir, topdown=False):
    for dir_name in dirs:
        if " " in dir_name:
            old_path = os.path.join(root, dir_name)
            new_dir_name = dir_name.replace(" ", "_")
            new_path = os.path.join(root, new_dir_name)

            if os.path.exists(new_path):
                print(f"이미 존재: {new_path}")
                continue

            os.rename(old_path, new_path)
            print(f"폴더명 변경: {dir_name} → {new_dir_name}")
```

```python
import os
import shutil

root_dir = r"C:\potenup3\potenup3-pj02-Project\서식\[별지+제1호서식]+확정일자+신청서(상가건물+임대차계약서상의+확정일자+부여+및+임대차+정보제공에+관한+규칙)\오답"  # 여기만 수정

files = os.listdir(root_dir)

for file in files:
    if file.lower().endswith(".pdf") and "-" in file:

        try:
            # 009-01.pdf → 01 추출
            suffix = file.split("-")[-1].split(".")[0]

            if not suffix.isdigit():
                continue

            source_path = os.path.join(root_dir, file)
            target_folder = os.path.join(root_dir, suffix)
            target_path = os.path.join(target_folder, file)

            # 폴더 생성
            os.makedirs(target_folder, exist_ok=True)

            # 이동
            shutil.move(source_path, target_path)

            print(f"{file} → {suffix}/ 이동 완료")

        except Exception as e:
            print(f"에러: {file} → {e}")
```

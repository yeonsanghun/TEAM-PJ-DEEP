```python
import os
import pandas as pd

# 🔹 정확한 루트 경로 (r"" 꼭 사용)
ROOT = r"C:\potenup3\TEAM-PJ-DEEP\document_forms_data\[별지+제1호서식]+확정일자+신청서(상가건물+임대차계약서상의+확정일자+부여+및+임대차+정보제공에+관한+규칙)"

LABELS = [
    "01_임차인_성명",
    "02_임차인_주민등록번호",
    "03_임차인_상호",
    "04_임차인_사업자등록번호",
    "05_임차인_주소",
    "06_임차인_전화번호",
    "07_임차인_휴대전화번호",
    "08_임대인_성명",
    "09_임대인_주민_등록번호",
    "10_임대인_주소",
    "11_임대인_전화번호",
    "12_임대인_휴대_전화_번호",
    "13_임대차_상가_건물_소재지",
    "14_임대차_계약일",
    "15_임대차_임대차기간",
    "16_임대차_보증금",
    "17_임대차_차임",
    "18_임대차_면적",
    "19_위임_성명",
    "20_위임_주민등록번호",
    "21_위임_신청인과의_관계",
    "22_위임_전화번호",
    "23_년_월_일",
    "24_신청인",
    "25_위임받은_자",
]


def clean_label(label):
    return label.replace("없음", "").rstrip("_")


def create_csv(split):
    images_dir = os.path.join(ROOT, split, "images")

    if not os.path.exists(images_dir):
        print(f"❌ {images_dir} 경로가 존재하지 않습니다.")
        return

    print(f"📂 사용 경로: {images_dir}")

    rows = []

    for file in os.listdir(images_dir):
        if not file.lower().endswith((".jpg", ".png", ".jpeg", ".webp")):
            continue

        name = os.path.splitext(file)[0]

        if "-" not in name:
            continue

        _, label_part = name.split("-", 1)

        # 기본값 전부 1
        label_dict = {label: 1 for label in LABELS}

        # 00_정답이면 그대로 유지
        if label_part != "00_정답":
            cleaned = clean_label(label_part)
            if cleaned in LABELS:
                label_dict[cleaned] = 0

        row = {"image_path": os.path.join(split, "images", file)}
        row.update(label_dict)
        rows.append(row)

    df = pd.DataFrame(rows)
    df = df[["image_path"] + LABELS]

    save_path = os.path.join(ROOT, split, "labels.csv")
    df.to_csv(save_path, index=False, encoding="utf-8-sig")

    print(f"✅ {split}/labels.csv 생성 완료")


# 🔥 실행
for split in ["train", "val", "test"]:
    create_csv(split)

print("🎉 전체 CSV 생성 완료")
```

```python
import os
from pdf2image import convert_from_path

# 변환할 루트 디렉토리
root_dir = r"C:\potenup3\potenup3-pj02-Project\서식\별지_제1호서식_확정일자신청서"

# Poppler bin 경로 (PATH에 이미 추가했으면 None)
poppler_path = r"C:\potenup3\potenup3-pj02-Project\poppler-25.12.0\Library\bin"  # ← 설치 경로로 수정 / PATH 설정했으면 None

for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file.lower().endswith(".pdf"):
            pdf_path = os.path.join(root, file)

            try:
                # PDF → 이미지 리스트
                images = convert_from_path(
                    pdf_path,
                    dpi=300,
                    poppler_path=poppler_path,  # PATH 설정했다면 삭제 가능
                )

                # 페이지별로 JPG 저장
                for i, image in enumerate(images):
                    jpg_filename = f"{os.path.splitext(file)[0]}_{i+1}.jpg"
                    jpg_path = os.path.join(root, jpg_filename)
                    image.save(jpg_path, "JPEG")

                print(f"변환 완료: {pdf_path}")

            except Exception as e:
                print(f"에러 발생: {pdf_path} → {e}")
```

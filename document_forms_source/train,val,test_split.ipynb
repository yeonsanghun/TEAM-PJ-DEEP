```python
import os
import shutil

root_dir = r"C:\potenup3\potenup3-pj02-Project\서식\[별지+제1호서식]+확정일자+신청서(상가건물+임대차계약서상의+확정일자+부여+및+임대차+정보제공에+관한+규칙)"

# 목적지 images 폴더 경로
train_images = os.path.join(root_dir, "train", "images")
val_images = os.path.join(root_dir, "val", "images")
test_images = os.path.join(root_dir, "test", "images")

# 폴더 생성
os.makedirs(train_images, exist_ok=True)
os.makedirs(val_images, exist_ok=True)
os.makedirs(test_images, exist_ok=True)

for root, dirs, files in os.walk(root_dir):

    # 이미 생성한 train/val/test 내부는 탐색 제외
    if any(x in root for x in ["\\train\\", "\\val\\", "\\test\\"]):
        continue

    for file in files:
        if not file.lower().endswith(".jpg"):
            continue

        source_path = os.path.join(root, file)

        # 파일명 기준 분류
        if file.startswith(("005", "010")):
            target_path = os.path.join(test_images, file)
        elif file.startswith("009"):
            target_path = os.path.join(val_images, file)
        else:
            target_path = os.path.join(train_images, file)

        # 이미 존재하면 스킵
        if os.path.exists(target_path):
            print(f"이미 존재: {file}")
            continue

        shutil.copy2(source_path, target_path)
        print(f"{file} → {os.path.relpath(target_path, root_dir)} 복사 완료")

print("✅ 분류 완료")
```
